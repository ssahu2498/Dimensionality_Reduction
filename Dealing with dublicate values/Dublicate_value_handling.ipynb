### Importing the libraries

```python
import pandas as pd
import numpy as np
```

### Creating the Employee dataframe 

```python
df_employees = pd.DataFrame([('Shubham', 25, 'Varanasi'), 
            ('Neha', 24, 'Ranchi'), 
            ('Aaditya', 25, 'Mumbai'), 
            ('Saumya', 32, 'Delhi'),
            ('Saumya', 32, 'Delhi'),
            ('Saumya', 32, 'Delhi'), 
            ('Saumya', 32, 'Mumbai'), 
            ('Aaditya', 40, 'Dehradun'), 
            ('Seema', 32, 'Delhi'),
             ('Shubham', 25, 'Varanasi'),
            ('Shubham', 25, 'Indore')
            ],columns = ['Name','Age',"City"])
df_employees 
```

### Checking the dublicate values in dataframe

```python
## We are using pandas.duplicated() method
df_employees.duplicated()
```

```python
#printing the duplicates rows here keep = 'first' means first similar rows are considered as unique not duplicate
df_employees[df_employees.duplicated(keep='first')]
```

```python
# For treating the last similar row as unique and rest  as duplicate
df_employees[df_employees.duplicated(keep='last')]
```

```python
# For treating all the similar row as duplicate
df_employees[df_employees.duplicated(keep=False)]
```

```python
## Finding duplicate values on the basis of particular column City: use subset argument.
df_employees[df_employees.duplicated(subset = ['City'])]
```

```python
## Finding duplicate values on the basis of particular column Name and keeping first similar row as unique
df_employees[df_employees.duplicated(subset = ['Name'])]
```

```python
## Finding duplicate values on the basis of particular column Name and City and treating all similar rows as duplicate
df_employees[df_employees.duplicated(subset = ['Name','City'],keep=False)]
```

#### Handling Duplicate values

```python
# We are keeping the original dataset safe as we are performing the actions with in the dataset
df_employees_copy = df_employees
```

```python
# Removing the duplicate values by using pandas_dataframe.drop_duplicates() methods
df_employees_copy.drop_duplicates()
```

```python
# dropping all the duplicate rows except last similar row conside as unique
df_employees_copy.drop_duplicates(keep='last')
```

```python
# dropping all the duplicate rows on the basis of a particular column except last similar row conside as unique
df_employees_copy.drop_duplicates(subset = ['City'],keep='last')
```

```python
# dropping all the duplicate all rows on the basis of a particular column 
df_employees_copy.drop_duplicates(subset = ['City'],keep=False)
```

```python
# dropping all the duplicate all rows on the basis of a particular column 
df_employees_copy.drop_duplicates(subset = ['Name'],keep=False)
```

```python
## Dropping all the duplicate rows
df_employees_copy.drop_duplicates(keep=False)
```
